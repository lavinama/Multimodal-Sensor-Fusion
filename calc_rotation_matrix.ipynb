The angle is **positive** when one of the following is true:

$X → Y$

$Y → Z$

$Z → X$

The angle is **negative** when one of the following is true:

$Y → X$

$Z → Y$

$X → Z$

```python
# Initial frame
# z is pointing outward
# y is pointing down
# x is pointing left

# Target frame
# z is pointing left
# y is pointing inward
# x is pointing up
```

```python
# Rotation Z-axis
import numpy as np
import math
theta_z = - np.pi/2

# yaw = Rotation_z_axis
Rotation_z_axis = np.array([[math.cos(theta_z), - math.sin(theta_z), 0],
                            [math.sin(theta_z),   math.cos(theta_z), 0],
                            [0,                        0           , 1]])
print(np.around(Rotation_z_axis))
```

```python
# Rotation Y-axis
import numpy as np
import math
theta_y = 0

# pitch = Rotation_y_axis
Rotation_y_axis = np.array([[math.cos(theta_y),  0, math.sin(theta_y)],
                            [0,                  1,                 0],
                            [-math.sin(theta_y), 0, math.cos(theta_y)]])
print(np.around(Rotation_y_axis))
```

```python
# Rotation X-axis
import numpy as np
import math
theta_x = - np.pi/2

# roll = Rotation_x_axis
Rotation_x_axis = np.array([[1,                 0,                  0],
                            [0, math.cos(theta_x), -math.sin(theta_x)],
                            [0, math.sin(theta_x), math.cos(theta_x)]])
print(np.around(Rotation_x_axis))
```

```python
# 3D coordinate from target frame

orig_coord = np.array([[21],
                       [2.5],
                       [0.9]])

# First rotate along the x axis
target_coord = np.matmul(Rotation_x_axis, orig_coord)
# print(np.around(target_coord, decimals=2))

# Second rotate along the y axis
target_coord = np.matmul(Rotation_y_axis, target_coord)
# print(np.around(target_coord, decimals=2))

target_coord = np.matmul(Rotation_x_axis, np.matmul(Rotation_y_axis, np.matmul(Rotation_z_axis, orig_coord)))
target_coord = np.matmul(Rotation_x_axis, np.matmul(Rotation_y_axis, Rotation_z_axis))

print("Rotation_z_axis")
print(np.around(Rotation_z_axis))
print("Rotation_x_axis")
print(np.around(Rotation_x_axis))
print("trans_matrix")
trans_matrix = np.matmul(Rotation_x_axis, Rotation_z_axis)
print(np.around(trans_matrix))
```
